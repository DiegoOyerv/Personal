# Limpiar los datos 

```python
NeoAnemias = Anemias.drop(Anemias.columns[-7:], axis=1)
```

```python
NeoAnemias.isnull().sum()
```

```python

# Reemplaza nulos en todas las columnas que comiencen con "DE","DX","CX"
for col in NeoAnemias.columns:
    if col.startswith(("DE", "DX","CX")):
        NeoAnemias[col] = NeoAnemias[col].fillna("no hubo")
```

```python
NeoAnemias.isnull().sum()
```

```python
NeoAnemias_SinRepetir = NeoAnemias.drop_duplicates(subset='ARC_REG', keep='first')
```
